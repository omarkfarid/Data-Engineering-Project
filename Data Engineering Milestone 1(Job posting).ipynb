#### Data Cleaning

```python
''' Helper Functions to ease the use of data cleaning built-in methods to
    1- Remove column name spaces and replace them with '_'
    2- Text remover to remove unwanted text or characters
    3- Text Splitter to split the text based on a certain character
    4 - Text Replace to replace characters
    5- Multiple null remover to remove nulls accross multiple columns based on a threshhold (number of missing values in coilumns) that you decide and pass to the function
 '''

def columns_spaces (df):
    df.columns = df.columns.str.replace(' ','_')

def text_remover (df,column,chars):
    for c in chars:
        df[column] = df[column].apply(lambda x : str(x).replace(c,''))

def text_split (df, column, split_char, indx):
    df[column] = df[column].apply(lambda x : str(x).split(split_char)[indx])

def text_replace (df, column, char1, char2):
    df[column] = df[column].apply(lambda x : str(x).replace(char1,char2))

def multiple_null_remover (df, threshhold):
    null_count = df.isnull().sum(axis=1)
    targeted_rows = df[null_count >= threshhold]
    df.drop(targeted_rows.index ,inplace = True)
```

#### Job Title Feature Engineering

```python
df_clean.Job_Title.unique()
```

```python
## Mappings to create 2 new columns that include the Job Type and the level of the required Candidate

title_mapping = {'data scientist': 'Data Science',
                'data engineer':'Data Engineering',
                'machine learning':'AI / Machine Learning',
                'manager':'Management',
                'director': 'Management',
                'president': 'C-Level'}

level_mapping = {'sr':'Senior',
                 'senior':'Senior',
                'jr':'Junior',
                 'junior': 'Junior',
                'lead':'Lead',
                'Principal':'Principal'}
```

```python
'''Functions that apply the mappings for the job type and job level mapping'''

def job_type(title,title_mapping):
    for key, value in title_mapping.items():
        if key in title.lower():
            return value
        else:
            continue
    return 'Other'

def job_level(title,level_mapping):
    for key, value in level_mapping.items():
        if key in title.lower():
            return value
        else:
            continue
    return 'Associate'
```

```python
df_clean['Job_Type'] = df['Job Title'].apply(job_type,title_mapping = title_mapping)
df_clean['Job_Level'] = df['Job Title'].apply(job_level,level_mapping = level_mapping)
```

**Creating the LookUp Table**

```python
import csv

def create_csv_lookup_table(file_path, data, column_1, column_2):
    # Open the CSV file in write mode
    with open(file_path, 'a', newline='') as csv_file:
        # Create a CSV writer object
        csv_writer = csv.writer(csv_file)

        # Write the header
        csv_writer.writerow([column_1, column_2])

        # Write the data
        for size_range, average_value in data.items():
            csv_writer.writerow([size_range, average_value])

        csv_writer.writerow([])

# Specify the file path to save the CSV file
csv_file_path = 'lookup_table.csv'

# Call the function to create and write to the CSV file
create_csv_lookup_table(csv_file_path, mapping, column_1 = 'Company Size Range', column_2 = 'Founded Year Average')
create_csv_lookup_table(csv_file_path, sector_map, column_1 = 'Sector', column_2 = 'Revenue')
create_csv_lookup_table(csv_file_path, title_mapping, column_1 = 'Title', column_2 = 'Job Type')
create_csv_lookup_table(csv_file_path, level_mapping, column_1 = 'Title', column_2 = 'Job Level')
```
